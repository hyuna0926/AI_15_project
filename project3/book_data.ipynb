```python
from sklearn.model_selection import train_test_split

target = 'star_line'

train, test = train_test_split(total_book, test_size=0.2, random_state=42)
train, val = train_test_split(train, test_size=0.2, random_state=42)

train.shape, test.shape, val.shape
```

```python
features = total_book.drop([target, 'star', 'title','author'], axis=1).columns

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
y_test = test[target]
```

```python
X_train.shape, y_train.shape, X_val.shape, y_test.shape
```

```python
#기준모델
from sklearn.metrics import accuracy_score

base = y_train.mode()[0]
baseline = len(y_train)*[base]
baseline_acc = accuracy_score(y_train, baseline)
print('기준모델의 정확도:', baseline_acc.round(3))
```

```python
from category_encoders import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats.distributions import uniform
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
import numpy as np

from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
rf =make_pipeline(
    OrdinalEncoder(),
    SimpleImputer(),
    RandomForestClassifier(
        n_estimators=200, random_state=42, class_weight='balanced',
        n_jobs=-1, max_depth=6, max_samples=40
    )
)

rf.fit(X_train, y_train)

print('훈련정확도', rf.score(X_train, y_train))
print('검증정확도', rf.score(X_val, y_val))

y_pred_proba = rf.predict_proba(X_val)[:,1]
print('roc_auc', roc_auc_score(y_val, y_pred_proba))

y_val_pred = rf.predict(X_val)
print(classification_report(y_val,y_val_pred))
```
