```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
def get_page(page_url):
    page = requests.get(page_url)
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup, page
```

```python
#007 romance 로맨스
#제목
def title(i,page_num):
    url = f"http://www.yes24.com/24/Category/Display/00100104601100{i}?FetchSize=40&ParamSortTp=05&PageNumber={page_num}"
    
    soup, page = get_page(url)
    list=[]
    for i in range(0,80):
        list.append(soup.select('div.goods_name>a')[i].text)
        title1 = list[0::2]
    return title1


#가격
def price(i,page_num):
    url = f"http://www.yes24.com/24/Category/Display/00100104601100{i}?FetchSize=40&ParamSortTp=05&PageNumber={page_num}"
    
    soup, page = get_page(url)

    price = soup.select('div.goods_price>em.yes_b')
    price_1=[]

    for r in range(len(price)):
        b = float(re.sub(",","",price[r].text))
        price_1.append(b)
    return price_1


#별점
def star(i, page_num):
    url = f"http://www.yes24.com/24/Category/Display/00100104601100{i}?FetchSize=40&ParamSortTp=05&PageNumber={page_num}"
    
    soup, page = get_page(url)
    star1=[]

    star =soup.select('span.gd_rating > em.yes_b')

    for r in range(len(star)):
        c = float(star[r].text)
        star1.append(c)
    return star1

#저자/역자
def author(i,page_num):
    url = f"http://www.yes24.com/24/Category/Display/00100104601100{i}?FetchSize=40&ParamSortTp=05&PageNumber={page_num}"
    
    soup, page = get_page(url)
    author1=[]
    for i in range(1,41):
        d =re.sub("\n|\r","",soup.select('span.goods_auth')[i].text)
        author1.append(d)
    return author1
    


#페이지별
def get(i, page_num):
    url = f"http://www.yes24.com/24/Category/Display/00100104601100{i}?FetchSize=40&ParamSortTp=05&PageNumber={page_num}"
    
    soup, page = get_page(url)

    star_1 =[]
    title_1 =[]
    price_1 =[]
    author_1 = []
    for page_n in range(page_num):
        star_1 +=star(i, page_n)
        title_1 +=title(i, page_n)
        price_1 += price(i, page_n)
        author_1 +=author(i, page_n)
    return star_1, title_1, price_1, author_1
```

```python
#리뷰수
def review(i,page_num):
    url = f"http://www.yes24.com/24/Category/Display/00100104601100{i}?FetchSize=40&ParamSortTp=05&PageNumber={page_num}"
    
    soup, page = get_page(url)
    review1=[]
    review_count = soup.select('span.gd_reviewCount > em.txC_blue')
    for i in range(len(review_count)):
        d =int(review_count[i].text)
        review1.append(d)
    return review1

#페이지별
def get(i, page_num):
    url = f"http://www.yes24.com/24/Category/Display/00100104601100{i}?FetchSize=40&ParamSortTp=05&PageNumber={page_num}"
    
    soup, page = get_page(url)

    review_1 = []
    for page_n in range(page_num):
        review_1 +=review(i, page_n)
    return review_1
```
